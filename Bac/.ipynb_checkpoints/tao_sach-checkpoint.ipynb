```python
text = """
“You’re playing to help me! Off you go,” said the lion amused as the little mouse was running away.

A few days go by when the little mouse heard a roar: “Help me!”

The lion was scaring but it walked on and on until it came across the lion caught in the net!

“I see you need my help now!” it said.

“Please help me!” groaned the lion in pain.

The little one immediately chewed across the net until the lion was free.

From that day on the lion and the mouse remained friends!
"""

time_sleep = 8


import csv
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.action_chains import ActionChains
import tkinter
from tkinter.font import Font
import time

page_url = "https://parts-of-speech.info/"
driver = webdriver.Chrome()
driver.get(page_url)

inputElement = driver.find_element(By.ID, 'text')
inputElement.clear()
inputElement.send_keys(text)
driver.find_element(By.ID, 'submit').click()

print("Giờ chạy tiếp khối lệnh thứ 2")

time.sleep(time_sleep)

words = []
tags = []
sub_tags = []

elements = driver.find_element(By.ID, 'textTagged')
elements = elements.find_elements(By.XPATH, './/*')

for i in range(len(elements)):
    words.append(elements[i].text)
    tags.append(elements[i].get_attribute('class').split()[-1][3:])
    hover = ActionChains(driver).move_to_element(elements[i])
    hover.perform()
    sub_tags.append(driver.find_element(By.ID, 'tagTip').text.split()[-1])

tkinter.Frame().destroy()

font1 = Font(family='EB Garamond', size=11)
font2 = Font(family='EB Garamond', size=9)
font3 = Font(family='EB Garamond', size=8)

W = 384

def random(*args):
    noun_indexes, verb_indexes, adj_indexes, adv_indexes= args
    
    n_noun = min(len(noun_indexes), 4)
    n_verb = min(len(verb_indexes), 6)
    n_adj = min(len(adj_indexes), 4)
    n_adv = min(len(adv_indexes), 2)
    
    choice_idx = []
    max_loop = 100
    loop_count = 0
    for idx_set, num in zip([noun_indexes, verb_indexes, adj_indexes, adv_indexes], [n_noun, n_verb, n_adj, n_adv]):
        c = 0
        while c < num:
            idx = np.random.choice(idx_set)
            if len(choice_idx) > 0 and np.min(np.abs(np.array(choice_idx) - idx)) < 4:
                loop_count += 1
                if loop_count > max_loop:
                    print('Tìm lại lần nữa!')
                    return random(*args)
                continue
            choice_idx.append(idx)
            c += 1
    print('Đã tìm thấy')
    return choice_idx

def sample(words, tags, sub_tags):
    noun_indexes = [i for i in range(len(tags)) if tags[i] == 'Noun' and len(words[i]) > 1 and words[i].islower()]
    verb_indexes = [i for i in range(len(tags)) if tags[i] == 'Verb' and len(words[i]) > 1]
    adj_indexes = [i for i in range(len(tags)) if tags[i] == 'Adjective' and len(words[i]) > 1]
    adv_indexes = [i for i in range(len(tags)) if tags[i] == 'Adverb' and len(words[i]) > 1]

    choice_idx = random(noun_indexes, verb_indexes, adj_indexes, adv_indexes)

    words = words[:]
    tags = tags[:]
    sub_tags[:]

    for i in range(len(words)):
        if i in choice_idx:
            space_length = len(words[i]) * 2
            words[i] = '_' * space_length
            if sub_tags[i] == "NNS": tags[i] = 'Plural noun'
            elif sub_tags[i] == "VBG": tags[i] = 'Verb ending in "ing"'
        else:
            tags[i] = ''

    return words, tags


def gen_txt(words, tags, W):
    
    s_words, s_tags = sample(words, tags, sub_tags)

    lines = ['']
    blanks = ['']
    no_space_after = ['“', '’', '(']
    no_space_before = [':', ')', '”', '?', '!', '.', ',', ';']
    for i in range(len(words)):
        if (len(words[i]) == 1 and not words[i].isalpha() and words[i] not in no_space_after) or words[i - 1] in no_space_after or '’' in words[i]:
            space = ''
        else:
            space = ' '
        if font1.measure(lines[-1] + space + s_words[i]) > W and s_words[i] not in no_space_before:
            lines.append('')
            blanks.append('')
            space = ''
        
        if s_tags[i] != '':
            center = font1.measure(lines[-1] + space) + font1.measure(s_words[i]) // 2
            space_num = (2 * (center - font2.measure(blanks[-1]) - font2.measure(s_tags[i])//2) // font2.measure(' ') + (center - font3.measure(blanks[-1]) - font3.measure(s_tags[i])//2) // font3.measure(' ')) // 3
            blanks[-1] = blanks[-1] + ' ' * space_num + s_tags[i]
            
        lines[-1] = lines[-1] + space + s_words[i]
            
    text = ''
    for i in range(len(lines)):
        l, b = lines[i], blanks[i]
        if i == 0:
            l = l[1:]
            b = b[1:]
        text += l + '\n' + b + '\n'
    text = text

    menu = '\n'.join([t.strip() + '_' * ((W - 2 - font1.measure(t.strip())) // font1.measure('_')) for t in s_tags if t != ''])
    
    return menu, text


menu, text = gen_txt(words, tags, W=W)

with open('file.txt', 'w') as f:
    f.write('\n\n'.join([menu, text]))

f.close()
driver.close()

print('OK')
```
