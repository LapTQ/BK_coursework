```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('Picture1.png')
img2 = cv2.imread('Picture2.png')

img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

img2 = cv2.resize(img2, (209, 169))

plt.imshow(img1, cmap='gray')
plt.show()
plt.imshow(img2, cmap='gray')
plt.show()
```

```python
# Process to have an image which has only the “ball”
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img1, cmap='gray')
plt.title('img1')
plt.subplot(1, 2, 2)
plt.imshow(img2, cmap='gray')
plt.title('img2')
plt.show()

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(255-img1, cmap='gray')
plt.title('not img1')
plt.subplot(1, 2, 2)
plt.imshow(255-img2, cmap='gray')
plt.title('not img2')
plt.show()

b = np.logical_and(img1, img2).astype(np.uint8)
plt.imshow(b, cmap='gray')
plt.title('b = img1 and img2')
plt.show()

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
c1 = np.logical_or(255-img1, b).astype(np.uint8)
plt.imshow(c1, cmap='gray')
plt.title('c1 = b or (not img1)')
plt.subplot(1, 2, 2)
c2 = np.logical_or(255-img2, b).astype(np.uint8)
plt.imshow(c2, cmap='gray')
plt.title('c2 = b or (not img2)')
plt.show()

result = np.logical_and(c1, c2).astype(np.uint8)
plt.imshow(result, cmap='gray')
plt.title('c1 and c2')
plt.show()
```
